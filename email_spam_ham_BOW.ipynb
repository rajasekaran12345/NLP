```python
import re
from nltk.stem import WordNetLemmatizer
import pandas as pd
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/bigmlcom/python/refs/heads/master/data/spam.csv',sep='\t')
df.head()
```

```python
from nltk.corpus import stopwords
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python

lemma=WordNetLemmatizer()
corpus=[]
for i in range(0,len(df)):
  review=re.sub('[^A-Za-z]',' ',df['Message'][i])
  review=review.lower()
  review=review.split()
  review=[lemma.lemmatize(word) for word in review if word not in set(stopwords.words('english'))]
  review=" ".join(review)
  corpus.append(review)
```

```python
corpus[:5]
```

```python
#creating bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=2500,ngram_range=(1,2))
```

```python
X=cv.fit_transform(corpus)
```

```python
cv.vocabulary_
```

```python
y=pd.get_dummies(df['Type'])
```

```python
y=y.iloc[:,:1].values
```

```python
y[1:6]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.20)
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detector_model=  MultinomialNB().fit(x_train,y_train)
```

```python
y_pred=spam_detector_model.predict(x_test)
```

```python
# metrics = accuracy score,classification report
from sklearn.metrics import accuracy_score,classification_report
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```
