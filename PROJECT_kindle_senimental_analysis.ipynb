```python
df=df[['reviewText','rating']]
```

```python
##preprocessing and cleaning
''' we are converting the rating categories such that rating
  less than 3 will be zero(positive) 3 or more than 3 is negative'''
df['rating']=df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
df['rating'].unique()
```

```python
## 1. lower all the cases
df['reviewText'].str.lower()
```

```python
import nltk
nltk.download('stopwords')
```

```python
from bs4 import BeautifulSoup
```

```python
'''removing the special characters'''
import re
df['reviewText'].apply(lambda x:re.sub('[^a-z A-Z 0-9]+'," ",x))
'''removing the stopwords'''
from nltk.corpus import stopwords
df['reviewText'].apply(lambda x:" ".join([y for y in x.split() if y not in stopwords.words("english")]))
url_pattern = r'http\S+|www\.\S+'
df['reviewText'] = df['reviewText'].apply(lambda x: re.sub(url_pattern, '', x))
''' remonving html tags using beautiful soup libray'''
df['reviewText'] = df['reviewText'].apply(lambda x: BeautifulSoup(x, "html.parser").get_text() if pd.notnull(x) else x)
'''Remove any additional spaces'''
df['reviewText']=df['reviewText'].apply(lambda x: " ".join(x.split()))
```

```python
df['reviewText'].head()
```

```python
import nltk
nltk.download('wordnet')
```

```python
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()

def lemma(x):
  return " ".join([lemmatizer.lemmatize(i) for i in x.split()])
```

```python
df['reviewText']=df["reviewText"].apply(lambda x:lemma(x))
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
bow=CountVectorizer()
tfidf=TfidfVectorizer()
x_train_bow=bow.fit_transform(x_train).toarray()
x_train_tfidf=tfidf.fit_transform(x_train).toarray()
x_test_bow=bow.transform(x_test).toarray()
x_test_tfidf=tfidf.transform(x_test).toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
model_bow=MultinomialNB()
model_tfidf=MultinomialNB()
model_bow.fit(x_train_bow,y_train)
model_tfidf.fit(x_train_tfidf,y_train)
```

```python
print(model_bow.score(x_test_bow,y_test))
print(model_tfidf.score(x_test_tfidf,y_test))
```

```python
def find_sentiment():
  review=input("give your review")
  review=" ".join([i for i in review.split() if i not in stopwords.words('english')])#returns a string
  review=re.sub(r'[^a-z A-Z 0-9]'," ",review)
  review=lemma(review)
  bow_review=bow.transform([review]).toarray()#tranform fuctions always epects a list of arrays
  pred=model_bow.predict(bow_review)
  if pred==1:
    return "it is a POSITVE REVIEW"
  else:
    return "it is a NEGATIVE REVIEW"

```

```python
'''predicting the sentiment of an output'''
find_sentiment()
```
